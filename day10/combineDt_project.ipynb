```python
import pandas as pd
import numpy as anp

students=pd.DataFrame({
    'StudentID':[1,2,3,4,5,6],
    'Name':['Aryan','Priya','Riya','Sam','Neha','Dev'],
    'Class': ['A','A','B','B','A','B'],
})

mrks_term1=pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5, 6],
    'Math': [85, 72, 91, 60, 88, 55],
    'Science': [78, 85, 70, 65, 92, 60],
})

mrks_term2=pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5, 6],
    'Math': [90, 75, 88, 64, 91, 58],
    'Science': [82, 88, 74, 68, 95, 63],
})

mrks_term1['Term']='Term1'
mrks_term2['Term']='Term2'
marks_wide=pd.concat([mrks_term1,mrks_term2], ignore_index=True)
print(marks_wide)
```

```python
mrks_long=marks_wide.melt(
    id_vars=['StudentID','Term'],value_vars=['Math','Science'],
    var_name='Subject',value_name='Marks')
print(mrks_long)
```

```python
full=pd.merge(mrks_long,students,on='StudentID',how='left')
print("Combined long table")
print(full.head(8))
```

```python
summary = full.pivot_table(
    index=['Class', 'Term'], columns='Subject',
    values='Marks', aggfunc='mean'
).round(1)
print()
print('=== AVERAGE MARKS BY CLASS & TERM ===')
print(summary)
 
# 6. Improvement column: Term2 - Term1 per student per subject (pivot back to wide)
wide_by_term = full.pivot_table(
    index=['StudentID', 'Name', 'Subject'], columns='Term', values='Marks'
).reset_index()
wide_by_term['Improvement'] = wide_by_term['Term2'] - wide_by_term['Term1']
print()
print('=== IMPROVEMENT TERM1 -> TERM2 ===')
print(wide_by_term)
 
# 7. Save combined outputs
full.to_csv('all_marks_combined.csv', index=False)
summary.to_csv('class_term_summary.csv')
```
