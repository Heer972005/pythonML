```python
#sectionA
import pandas as pd
employees = pd.DataFrame({
'EmpID': [101, 102, 103, 104, 105],
'Name': ['Asha', 'Rohan', 'Meera', 'Karan', 'Tina'],
'Department': ['HR', 'IT', 'IT', 'Finance', 'HR'],
})
salaries = pd.DataFrame({
'EmpID': [101, 102, 103, 106],
'Salary': [45000, 60000, 55000, 50000],
})
```

```python
#Q1
merges=pd.merge(employees,salaries,on='EmpID')
print(merges)

#Q2
merge_left=pd.merge(employees,salaries,on='EmpID',how='left')
print("\n",merge_left)

#Q3
merge_out=pd.merge(employees,salaries,on='EmpID',how='outer')
print("\n",merge_out)
```

```python
#section C

import pandas as pd
wide_scores = pd.DataFrame({
'Name': ['Asha', 'Rohan', 'Meera'],
'Quiz1': [8, 6, 9],
'Quiz2': [7, 8, 10],
'Quiz3': [9, 7, 8],
})
```

```python
#Q8
score_long=wide_scores.melt(
    id_vars=['Name'],var_name='Quiz',value_name='Score')
print(score_long)

#Q9
print("\n",score_long.pivot(index='Name',columns='Quiz',values='Score'))

#10
avg=score_long.groupby('Name')['Score'].mean()
print("\nStudent with highest average is",avg.idxmax(),"scoring",avg.max())

#Q11
avg_score_per_quiz=score_long.groupby('Quiz')['Score'].mean()
print("\n",avg_score_per_quiz)
print(avg_score_per_quiz.idxmin(),"has the lowest average score which is",avg_score_per_quiz.min())
```

```python
#secction D
import pandas as pd
records = pd.DataFrame({
'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
'Quarter': ['Q1', 'Q2', 'Q1', 'Q2', 'Q1', 'Q2'],
'Expense': [50000, 52000, 30000, 31000, 45000, 47000],
})
```

```python
#Q12
grouped=records.groupby(['Department','Quarter'])['Expense'].sum()
print(grouped)

#Q13
print(grouped.loc[('HR','Q2')])

#Q14
flat=grouped.reset_index()
print(flat)
print(flat.pivot(index='Department',columns='Quarter',values='Expense'))
```

```python
# Section E
import pandas as pd
students = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'Name': ['Asha', 'Rohan', 'Meera', 'Karan'],
'Section': ['A', 'A', 'B', 'B'],
})
scores_term1 = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'English': [70, 65, 80, 60],
'Maths': [85, 60, 75, 90],
})
scores_term2 = pd.DataFrame({
'StudentID': [1, 2, 3, 4],
'English': [75, 68, 82, 64],
'Maths': [88, 63, 78, 92],
})
```

```python
#Q15
scores_term1['Term']='Term1'
scores_term2['Term']='Term2'
all_scores_wide=pd.concat([scores_term1,scores_term2],ignore_index=True)
print(all_scores_wide)

#Q16
score_long=all_scores_wide.melt(id_vars=['StudentID','Term'],var_name='Subject',value_name='Score')
print(score_long)

#Q17
full=pd.merge(score_long,students,on='StudentID',how='left')
print("\nfull tble\n",full)

#Q18
summary=full.pivot_table(index=['Section','Term'],columns='Subject',values='Score',aggfunc='mean').round(1)
print("\n",summary)

#Q19
wide=full.pivot_table(index=['StudentID','Name','Subject'],columns='Term',values='Score').reset_index()
wide['Improvement'] = (
    wide['Term2'] - wide['Term1']
)
print("\n",wide)

#Q20
best_row = wide.loc[wide['Improvement'].idxmax()]

print(f"{best_row['Name']} has shown the largest improvement by {best_row['Improvement']} marks.")
print(wide.loc[wide['Improvement'].idxmax()])
```
