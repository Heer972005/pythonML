```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, StandardScaler

df = pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5, 6],
    'City': ['Rajkot', 'Mumbai', 'Pune', 'Rajkot', 'Delhi', 'Pune'],
    'Education': ['Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters', 'PhD'],
    'Age': [21, 24, 29, 35, 41, 47],
    'Score': [78, 85, 91, 65, 72, 88],
})

city_encoded = pd.get_dummies(df['City'], prefix='City')


edu_order = OrdinalEncoder(categories=[['Bachelors', 'Masters', 'PhD']])
df['Education_Encoded'] = edu_order.fit_transform(df[['Education']])
 

scaler = StandardScaler()
df[['Age_z', 'Score_z']] = scaler.fit_transform(df[['Age', 'Score']])


df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 25, 35, 100],
                         labels=['Young', 'Mid-career', 'Senior'])

final = pd.concat([df.drop(columns=['City']), city_encoded], axis=1)
print(final.head())


final.to_csv('students_feature_engineered.csv', index=False)
```
