```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

df=pd.DataFrame({'City':['Rajkot','Mumbai','Pune','Rajkot','Delhi']})
print(df)
one_hot=pd.get_dummies(df,columns=['City'])
print(one_hot)

encoder=OneHotEncoder(sparse_output=False)
encoded=encoder.fit_transform(df[['City']])
print(encoder.get_feature_names_out())
print(encoded)
```

```python
#creats binary columns for every category ---safest for default nominal data
#sparse matrix--A sparse matrix is a matrix where the vast majority of the elements are zero. 
# Because storing these empty spaces wastes massive amounts of memory, computers use specialized data structures and 
# storage formats to keep only the non-zero values and their coordinates

#fit=gets the unique names
#tranform=transforms into numerical

#dummies-pandas
#onehotencoder-sckit learn
```

```python
#label encoder
le=LabelEncoder()
df['City_Label']=le.fit_transform(df['City'])
print(df)
```

```python
#ordinal 
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df=pd.DataFrame({
    'Size':['S','M','L','XL','M','S'],
    'Education':['Bachelors','Masters','PhD','Bachelors','Masters','PhD']
})

encoder=OrdinalEncoder(categories=[['S','M','L','XL']])
df['Size_Encoded']=encoder.fit_transform(df[['Size']])
print(df[['Size','Size_Encoded']])

#manual mapping
education_order={'Bachelors':0,'Masters':1,'PhD':2}
df['Education_Encoded']=df['Education'].map(education_order)
print(df[['Education','Education_Encoded']])
```

```python
#for ordinal number --reala order--encoding should preserve it
```

```python
#Scaling
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler

df=pd.DataFrame({
    'Age':[22,25,47,35,60],
    'Income':[25000,32000,95000,58000,120000]
})

minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=['Age_mm', 'Income_mm'])
print(df_minmax.round(3))

# StandardScaler -- z-score standardization (mean=0, std=1)
standard = StandardScaler()
df_standard = pd.DataFrame(standard.fit_transform(df), columns=['Age_z', 'Income_z'])
print(df_standard.round(3))
```

```python
#MinMaxScaler----no maajor outliers ad need a bounded range
#(x−min)/(max−min) range from 0 to 1
#StandardScaler---normal distribution as it is snot bounded by data's min/max extremes
#(x−min)/(max−min)---after mean=0,sd=1
# Distance-based algorithms like:

# KNN
# K-Means
# PCA
# Neural Networks

# will pay much more attention to Income and ignore Age.

# Therefore we scale features.
```

```python
#binning continuous variables
#converting contunious numericaal to discrete

import pandas as pd
ages = pd.DataFrame({'Age': [12, 19, 25, 34, 45, 52, 67, 8, 29, 71]})
# pd.cut() -- equal-WIDTH bins
bins = [0, 18, 35, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
ages['AgeGroup'] = pd.cut(ages['Age'], bins=bins, labels=labels)
print(ages)
 
# pd.qcut() -- equal-FREQUENCY bins, based on quantiles
ages['AgeQuartile'] = pd.qcut(ages['Age'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
print("\n",ages[['Age', 'AgeQuartile']])
```

```python
#qcut---Pandas decides boundaries so that each group contains equal numbers of rows
```
