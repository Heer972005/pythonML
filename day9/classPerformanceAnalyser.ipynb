```python
import pandas as pd
import numpy as np

np.random.seed(1)
n=24
df=pd.DataFrame({
    'Name':[f'Student_{i}'for i in range(1,n+1)],
    'Class': np.random.choice(['A','B','C'],n),
    'Subject':np.random.choice(['Math','Science','Englisg'],n),
    'Marks':np.random.randint(40,100,n)
})

df.loc[0,"Marks"]=23
print(df)
class_summary=df.groupby('Class')['Marks'].agg(avg_marks='mean',highest='max',lowest='min',students='count').round(1)
print('='*3,'CLASS SUMMARY','='*3)
print(class_summary)

pivot=df.pivot_table(index='Class',columns='Subject',values='Marks',aggfunc='mean').round(1)
print()
print('CLASS x SUBJECCT PIVOT')
print(pivot)

df['Result']=np.where(df['Marks']>=40,'Pass','Fail')
print()
print("PASS/FAIL COUNT BY CLASS")
print(pd.crosstab(df['Class'],df['Result']))
```

```python
df['class_avg']=df.groupby('Class')['Marks'].transform('mean').round(1)
df['deviation']=(df['Marks']-df['class_avg']).round(1)
print()
print(df.head(10).to_string(index=False))

class_summary.to_csv('class_summary.csv')
df.to_csv('class_performance.csv',index=False)
```
