```python
import pandas as pd
import numpy as np 
np.random.seed(1) 
students = [f'Student_{i}' for i in range(1, 31)] 
classes  = np.random.choice(['A','B','C'], 30) 
subjects = ['Math','Science','English','ICT'] 
rows = [] 
for s, c in zip(students, classes): 
    for subj in subjects: 
        rows.append({ 
            'Student': s, 
            'Class'  : c, 
            'Subject': subj, 
            'Marks'  : np.random.randint(30, 100) 
        }) 
df = pd.DataFrame(rows) 
print(df.head(10)) 
print(f'Shape: {df.shape}')
```

```python
#Task 1
class_avg=df.groupby('Class')['Marks'].mean().round(2)
print(class_avg)
print(f'{class_avg.idxmax()} class has the highest average marks')
```

```python
#task 2
subject_avg=df.groupby('Subject')['Marks'].mean().round(1)
print(subject_avg.sort_values())
```

```python
#task 3
class_stats=df.groupby('Class')['Marks'].agg(class_avg="mean",min_mrks="min",max_mrks="max",num_records="count")
print(class_stats.reset_index())
```

```python
#task4
pivot=df.pivot_table(index='Class',columns='Subject',values='Marks',aggfunc='mean',margins=True,margins_name='Total').round(1)
print(pivot)
```

```python
#task 5
df['Result']=np.where(df['Marks']>=40,'Pass','Fail')
print(pd.crosstab(df['Class'],df['Result'],normalize='index').round(1))
```

```python
#task 6
df['class_avg']=df.groupby('Class')['Marks'].transform('mean').round(1)
df['deviation']=(df['Marks']-df['class_avg']).round(1)
print()
print(df.sort_values(by='deviation',ascending=False).head(10).reset_index())
```

```python
sales_data = { 
'Region'  : ['North','North','South','South','East','East','West','West'], 
'Product' : ['A','B','A','B','A','B','A','B'], 
'Units'   : [120, 85, 95, 110, 60, 75, 130, 90], 
'Revenue' : [2400, 1700, 1900, 2200, 1200, 1500, 2600, 1800], 
} 
sales = pd.DataFrame(sales_data) 
print(sales) 
```

```python
#task 7
total_revenue=sales.groupby('Region')['Revenue'].sum()
print(total_revenue)

product_avg_sold=sales.groupby('Product')['Units'].mean()
print(product_avg_sold)

pivot=sales.pivot_table(index='Region',columns='Product',values='Revenue',aggfunc='sum',margins=True,margins_name='Grand Totals')
print("\n",pivot)
```

```python
#task 8
sales['Revenue_per_Unit']=(sales['Revenue']/sales['Units']).round(2)
avg_revenue_per_unit=sales.groupby('Region')['Revenue_per_Unit'].mean().sort_values(ascending=False)
print(f'The most efficient region is {avg_revenue_per_unit.idxmax()} with an average revenue per unit of {avg_revenue_per_unit.max()}')
```
