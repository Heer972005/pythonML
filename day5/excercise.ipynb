```python
import numpy as np

np.random.seed(42)

arr = np.random.randint(50, 101, 10)

print("Array:")
print(arr)

print("\nStatistics")
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr, ddof=1))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

q1 = np.percentile(arr, 25)
q3 = np.percentile(arr, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = arr[(arr < lower) | (arr > upper)]

print("\nOutliers:")
print(outliers if len(outliers) > 0 else "No outliers found")
```

```python
import numpy as np

marks = np.array([
    [78, 85, 92],
    [65, 70, 68],
    [90, 88, 95],
    [55, 60, 58]
])

print("Marks Array:")
print(marks)

student_avg = np.mean(marks, axis=1)

subject_avg = np.mean(marks, axis=0)

print("\nStudent Averages:")
print(student_avg)

print("\nSubject Averages:")
print(subject_avg)
```

```python
import numpy as np

marks = np.array([82, 60, 45, 91, 75, 54, 67])

grades = np.where(
    marks >= 75,
    'A',
    np.where(
        marks >= 55,
        'B',
        'C'
    )
)

print("Marks:")
print(marks)

print("\nGrades:")
print(grades)
```

```python
import pandas as pd
import numpy as np

data = {
    'Math': [78, 85, 92, 65, 88],
    'Science': [72, 80, 90, 68, 85],
    'English': [75, 82, 89, 70, 84]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\nDescribe:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Math': [78, 85, 92, 65, 88],
    'Science': [72, 80, 90, 68, 85],
    'English': [75, 82, 89, 70, 84]
}

df = pd.DataFrame(data)

df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# Histogram

plt.figure(figsize=(6,4))
plt.hist(df['Average'], bins=5)
plt.title("Distribution of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Frequency")
plt.savefig("average_histogram.png")
plt.show()

# Box Plot

plt.figure(figsize=(6,4))
plt.boxplot(
    [df['Math'], df['Science'], df['English']],
    labels=['Math', 'Science', 'English']
)

plt.title("Subject-wise Marks Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.savefig("subject_boxplot.png")
plt.show()
```
