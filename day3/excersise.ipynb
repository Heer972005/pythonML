```python
num=10
user=int(input("enter the number"))
while user!=num:
    if user<num:
        print("Too low")
    else:
        print("Too high")
    user=int(input("enter the number"))
print("Congratulations! You guessed the number.")
```

```python
mrks=[72,55,91,0,88,45,63]
pss=0
fail=0
for i in mrks:
    if i>=40:
        print(f"{i} - Pass")
        pss+=1
    elif i==0:
        continue
    else:
        print(f"{i} - Fail")
        fail+=1
print(f"Total Pass: {pss}")
print(f"Total Fail: {fail}")
    
```

```python
totl=0
for i in range(1,6):
    mrks=int(input(f"Enter marks for subject {i}: "))
    totl=totl+mrks
avg=totl/5
print(f"Total marks: {totl}")
print(f"Average marks: {avg:.2f}")
```

```python
pro=1
for i in range(5,11):
    num1=i
    while(i>0):
        pro=pro*i
        i=i-1
    print(f"factorial of {num1} is {pro}")
```

```python
num=4
pro=1
while num>0:
    pro=pro*num
    num=num-1
print(f"factorial is {pro}")
```
