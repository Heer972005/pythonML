```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 1000

df = pd.DataFrame()

df['Study_Hours'] = np.random.normal(5, 2, n).clip(0, 12)
df['Attendance'] = np.random.normal(80, 12, n).clip(40, 100)
df['Sleep_Hours'] = np.random.normal(7, 1.5, n).clip(3, 10)
df['Previous_Grade'] = np.random.normal(65, 15, n).clip(20, 100)
df['Assignments_Completed'] = np.random.randint(0, 11, n)
df['Days_Absent'] = np.random.poisson(5, n)
df['Extracurricular_Hours'] = np.random.normal(4, 2, n).clip(0, 12)
df['Stress_Level'] = np.random.randint(1, 11, n)

score = (
      0.30 * df['Study_Hours']
    + 0.04 * df['Attendance']
    + 0.03 * df['Previous_Grade']
    + 0.20 * df['Assignments_Completed']
    - 0.20 * df['Days_Absent']
    - 0.15 * df['Stress_Level']
    + np.random.normal(0, 2.5, n)
)

prob = 1 / (1 + np.exp(-(score - 5)))
df['Pass'] = np.random.binomial(1, prob)
print(df['Pass'].value_counts())

df.to_csv(
    "students_feature_engineered_v3.csv",
    index=False
)
```
