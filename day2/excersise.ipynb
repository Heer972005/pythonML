```python
num1=input("Enter first number: ")
num2=input("Enter second number: ")
sign=input("Enter the operator (+, -, *, /, %): ")

if sign=="+":
     sum=float(num1)+float(num2)
     print(f"Sum: {sum:.2f}")
elif sign=="-":
     diff=float(num1)-float(num2)
     print(f"Difference: {diff:.2f}")
elif sign=="*":
     prod=float(num1)*float(num2)
     print(f"Product: {prod:.2f}")
elif sign=="/":
     quot=float(num1)/float(num2)
     print(f"Quotient: {quot:.2f}")
elif sign=="%":
     remainder=float(num1)%float(num2)
     print(f"Remainder: {remainder:.2f}")
else:
     print("Invalid operator")
```
