```python
marks = [85, 92, 78, 65, 88, 95, 72, 81]

total = 0
count = 0

for mark in marks:
    total += mark
    count += 1

average = total / count

print("Total (without built-in functions):", total)
print("Average (without built-in functions):", average)

# Verification using built-in functions
print("Total using sum():", sum(marks))
print("Average using built-in functions:", sum(marks)/len(marks))
```

```python
marks = [85, 92, 78, 65, 88, 95, 72, 35]

grades = [
    "A+" if mark >= 90 else
    "A" if mark >= 75 else
    "B" if mark >= 60 else
    "C" if mark >= 40 else
    "F"
    for mark in marks
]

print("Marks :", marks)
print("Grades:", grades)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Heer", "Riya", "Amit", "Jay", "Neha"],
    "Age": [19, 20, 18, 21, 19],
    "City": ["Rajkot", "Surat", "Ahmedabad", "Vadodara", "Jamnagar"],
    "Score": [92, 78, 65, 88, 55]
}

df = pd.DataFrame(data)

# Introduce missing values
df.loc[1, "Age"] = np.nan
df.loc[2, "City"] = np.nan
df.loc[4, "Score"] = np.nan
df.loc[3, "Name"] = np.nan

print("DataFrame with Missing Values:")
print(df)

print("\nMissing Values Count:")
print(df.isnull().sum())

# Fill numeric columns with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Score"] = df["Score"].fillna(df["Score"].mean())

# Fill text columns with Unknown
df["Name"] = df["Name"].fillna("Unknown")
df["City"] = df["City"].fillna("Unknown")

print("\nCleaned DataFrame:")
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Heer", "Riya", "Amit", "Jay", "Neha"],
    "Age": [19, np.nan, 18, 21, 19],
    "City": ["Rajkot", "Surat", np.nan, "Vadodara", "Jamnagar"],
    "Score": [92, 78, 65, 88, np.nan]
}

df = pd.DataFrame(data)

# Clean missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Score"] = df["Score"].fillna(df["Score"].mean())
df["Name"] = df["Name"].fillna("Unknown")
df["City"] = df["City"].fillna("Unknown")

# Add Passed column
df["Passed"] = df["Score"] >= 40

# Count passed students
passed_count = df["Passed"].sum()

print("Final DataFrame:")
print(df)

print("\nNumber of Students Passed:", passed_count)

# Save to CSV
df.to_csv("students.csv", index=False)

print("\nDataFrame saved to 'students.csv'")
```
