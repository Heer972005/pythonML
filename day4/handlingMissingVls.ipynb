```python
import pandas as pd
import numpy as np

dt={
    "Nme":["John","Jane","Doe",None,"Smith"],
    "Mrks":[85,np.nan,78,92,np.nan],
    "City":["New York","Los Angeles",None,"Chicago","Houston"]
}

df=pd.DataFrame(dt)
print(df.isnull())
print(df.isnull().sum()) #count missing per column
print(df.isnull().sum()/len(df)*100) #% missing values

df_dropped=df.dropna() #drop rows with missing values
df_dropped2=df.dropna(subset=["Mrks"]) #drop rows with missing values in "Mrks" column

df["Mrks"]=df["Mrks"].fillna(df["Mrks"].mean()) #fill missing values in "Mrks" column with mean
df["City"]=df["City"].fillna("Unknown") #fill missing values in "City" column with "Unknown"
df["Nme"]=df["Nme"].ffill() #forward fill missing values in "Nme" column

print(df)

```

```python
print(df.duplicated().sum())
print(df[df.duplicated()])
df=df.drop_duplicates() #drop duplicate rows
print(df)

df=df.drop_duplicates(
    subset=["Nme","Mrks"],#cconsider only these columns
    keep="first" #keep the first occurrence of duplicates
)
```

```python
print(df.dtypes)
df["Mrks"]=pd.to_numeric(df["Mrks"],errors="coerce") #convert to numeric, set non-convertible to NaN
df["Date"]=pd.to_datetime(df["Date"],format="%d/%m/%Y")
df["Grade"]=df["Grade"].astype(str)
df["Score"]=df["Score"].astype(float) #convert to datetime, set non-convertible to NaT
```

```python
df.rename(columns={"Nme":"Name","Mrks":"Marks"},inplace=True) #rename columns
df.columns=df.columns.str.lower().str.replace(" ","_")
df=df.reset_index(drop=True) #reset index after dropping rows
```
