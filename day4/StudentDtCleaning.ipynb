```python
import pandas as pd
import numpy as np
rwdt={
    "Nme": ["Alice", "Bob", "Charlie", "Alice", None, "Eve"],
    "Math":[85, np.nan, 95, 85, 92, "N/A"],
    "Science":[88, 90, 95, 88, 85, 88],
    "English":[90, 85, 88, 82, 91, 90]
}

df=pd.DataFrame(rwdt)
print("-"*5,"Raw DataFrame","-"*5)
print(df)
print(f"Shape:{df.shape}")
print(df[df.duplicated()])
df=df.drop_duplicates()
print(f"After dedup:{df.shape[0]}")

df["Math"]=pd.to_numeric(df["Math"],errors="coerce") #convert to numeric, set non-convertible to NaN
df["Nme"]=df["Nme"].fillna("Unknown") #fill missing names with "Unknown"
df[Math]=df["Math"].fillna(df["Math"].mean()) #fill missing math scores with mean
df["Total"]=df["Math"]+df["Science"]+df["English"] #add total score column
df["Avg"]=(df["Total"]/3).round(1) #add average score column
df["Grade"]=["A+" if a>=90 else "A" if a>=75 else "B" if a>=60 else "C" if a>=40 else "F" for a in df["Avg"]]


df=df.reset_index(drop=True) #reset index after dropping rows
print()
print("-"*5,"Processed DataFrame","-"*5)
print(df.to_string(index=False))
df.to_csv("processed_students.csv",index=False)
print("DataFrame saved to processed_students.csv")

'''
grades = []

for a in df["Avg"]:
    if a >= 90:
        grades.append("A+")
    elif a >= 75:
        grades.append("A")
    elif a >= 60:
        grades.append("B")
    elif a >= 40:
        grades.append("C")
    else:
        grades.append("F")

df["Grade"] = grades
'''
```
